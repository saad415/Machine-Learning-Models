# Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
```

# Import Data

```python
df = pd.read_csv( "salary_data.csv" )
df.head()
X = df.drop('Salary', axis=1).values
Y = df['Salary'].values

plt.scatter(X,Y)
```

# Train Model 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0 )
plt.scatter(X_train,Y_train, label='Training Data', color='g', alpha=.7)
plt.scatter(X_test,Y_test,  label='Test Data', color='r', alpha=.7)
plt.title("Test Train Split")



ml=LinearRegression()
ml.fit(X_train,Y_train)
```

# Test Model

```python
c = ml.intercept_
m = ml.coef_
#print(m)
#print(c)

Y_pred = ml.predict(X_train)
#print(Y_pred)
```

# Evaluate Model 

```python
r2 = r2_score(Y_train,Y_pred)
rmse = np.sqrt(mean_squared_error(Y_train,Y_pred))
print(r2)
print(rmse)
```

# Plot Results 

```python
#plt.scatter(X,Y)
#print(X_test)
#print(Y_pred)

#Y_pred
#Y_test
plt.scatter(X_train,Y_train)
plt.plot(X_train,Y_pred, label='Training Data', color='g', alpha=.7)
```
